```python
import pandas as pd
filepath = "TMDB_movie_dataset_v11.csv"
movies = pd.read_csv(filepath)
```

```python
movies.head()
```

```python
movies = movies[['id', 'title', 'overview', 'genres', 'vote_count', 'vote_average', 'poster_path', 'release_date']]
```

```python
movies.head()
```

```python
movies = movies.sort_values(by='vote_count', ascending=False)
```

```python
movies
```

```python
movies = movies.dropna()
movies.isnull().sum()
```

```python
movies
```

```python
movies = movies.head(10000)
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['genres']
```

```python
movies
```

```python
!pip install scikit-learn
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words="english")
cv
```

```python
vector = cv.fit_transform(movies['tags'].values.astype('U')).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
def recommend(film):
    index = movies[movies['title']==film].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:15]:
        print(movies.iloc[i[0]].title)
```

```python
recommend("The Avengers")
```

```python
import pickle
pickle.dump(movies, open('movies_list.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
pickle.load(open('movies_list.pkl', 'rb'))
```
